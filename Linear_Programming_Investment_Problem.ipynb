```python
from pulp import *

# Create the model
model = LpProblem("Investment_Planning", LpMaximize)

# Decision variables: amounts invested at the start of each year
# A, B, C, and D are different investment projects, respectively.

A1 = LpVariable("A1", lowBound=0)
A2 = LpVariable("A2", lowBound=0)
A3 = LpVariable("A3", lowBound=0)
A4 = LpVariable("A4", lowBound=0)

B1 = LpVariable("B1", lowBound=0)
B2 = LpVariable("B2", lowBound=0)
B3 = LpVariable("B3", lowBound=0)

C2 = LpVariable("C2", lowBound=0)
D5 = LpVariable("D5", lowBound=0)

# Idle cash carried over from start of year t to start of year t+1
R1 = LpVariable("R1", lowBound=0)
R2 = LpVariable("R2", lowBound=0)
R3 = LpVariable("R3", lowBound=0)
R4 = LpVariable("R4", lowBound=0)
R5 = LpVariable("R5", lowBound=0)

# Objective: maximize cash at beginning of year 6 (end of year 5)

model += 1.4*A4 + 1.7*B3 + 1.9*C2 + 1.3*D5 + R5

# Yearly cash-flow balance constraints (equalities)

model += A1 + B1 + R1 == 60000                        # Year 1
model += A2 + B2 + C2 + R2 == R1                      # Year 2
model += A3 + B3 + R3 == 1.4*A1 + R2                  # Year 3
model += A4 + R4 == 1.4*A2 + 1.7*B1 + R3              # Year 4
model += D5 + R5 == 1.4*A3 + 1.7*B2 + R4              # Year 5

# Solve
model.solve()

# Display results
print("Status:", LpStatus[model.status])
print(f"Optimal investment plan:")
print(f"A1 = {value(A1):.2f}") #  take the number value(A1) and format it as a floating‑point number with exactly 2 decimal places.
print(f"A2 = {value(A2):.2f}")
print(f"A3 = {value(A3):.2f}")
print(f"A4 = {value(A4):.2f}")
print(f"B1 = {value(B1):.2f}")
print(f"B2 = {value(B2):.2f}")
print(f"B3 = {value(B3):.2f}")
print(f"C2 = {value(C2):.2f}")
print(f"D5 = {value(D5):.2f}")
print(f"R1 = {value(R1):.2f}")
print(f"R2 = {value(R2):.2f}")
print(f"R3 = {value(R3):.2f}")
print(f"R4 = {value(R4):.2f}")
print(f"R5 = {value(R5):.2f}")
print(f"\nTotal cash at beginning of year 6: ${value(model.objective):.2f}")


# All idle cash (Rt) stays at zero because any idle money would miss investment returns.
# The model uses equality constraints because we force all available cash at each year to be either invested or carried forward as Rt (no “waste” allowed, but idle cash can be held and later invested). 
# The solver confirms that the best plan eliminates all idle cash.
```
