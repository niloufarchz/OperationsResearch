```python
from pulp import *
model = LpProblem('Whitt_Window', LpMaximize)
x1 = LpVariable("WoodFramed", 0) # Lowbound is zero, column based modelling isn't used
x2 = LpVariable("AluminumFramed", 0)

model += 300*x1 + 150*x2 # objective
model += 6*x1 + 8*x2 <= 48 # glass constraint
model += x1 <= 6 # aluminum frame limit
model += x2 <= 4

model.solve()
print(value(x1), value(x2), value(model.objective))
```

```python
# LpVariable(name, lowBound=None, upBound=None, cat='Continuous', e=None)
# lowBound — the smallest allowed value for the variable. Default is None, which means 0 (or actually unbounded in the sense of no lower bound, but by default PuLP variables are non‑negative unless you set lowBound=None and later treat them as free). When you write LpVariable("WoodFramed", 0), you’re explicitly setting lowBound=0, making it a standard non‑negative variable.
# upBound — the largest allowed value. Default None means no upper limit (infinity). If you wanted a variable that can be at most, say, 5, you’d write LpVariable("x", 0, 5).
# cat — the category: 'Continuous' (default), 'Integer', or 'Binary'. Since you left it out, your variables are continuous (fractional windows allowed, which matches the LP formulation).
# e — used for column generation; leave it out unless you’re doing advanced stuff.
# f you had tried to write something like LpVariable("WoodFramed", 0, None) or LpVariable("WoodFramed", 0, none) (lowercase) and got an error, it might be because None is a keyword and must be capitalized, or because you tried to pass None as a string "None". But simply omit it, which is the most common (and cleanest) way.
# In PuLP, an LpProblem object uses the += operator as a shorthand to add something to the model. You’re essentially “building” the model step by step, and each += adds one piece. It’s just PuLP’s syntax for addObjective / addConstraint. You don’t have to use +=, but it’s the idiomatic (and most readable) way.
```
